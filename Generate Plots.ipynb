```python
## Comparing the results between the original paper method and the new KKR method (fig 1): 

h_lst1 = [0.1, 0.5, 1, 2, 4, 8, 10]  # h in (a.u)
velocity_lst1 = np.linspace(0, 10, 30)
gamma1 = 0.1 

force_stp_dataset1 = []
force_im_dataset1 = []
force_stp_new_dataset1 = []
force_im_new_dataset1 = []

for index, h in enumerate(h_lst1):
    print('processing h = {}'.format(h))
    force_stp = []
    force_im = []
    force_stp_new = []
    force_im_new = []

    for v in velocity_lst1:
        fstped = base.f_stp(v, gamma1, h)
        fim = base.f_im(v, gamma1, h)
        fstped_new = base.f_stp_gamma0(v, h)
        fim_new = base.f_im_gamma0(v, h)
        force_stp.append(fstped)
        force_im.append(fim)
        force_stp_new.append(fstped_new)
        force_im_new.append(fim_new)

    force_im_dataset1.append(-1 * np.array(force_im))
    force_stp_dataset1.append(-1 * np.array(force_stp))
    force_im_new_dataset1.append(-1 * np.array(force_im_new))
    force_stp_new_dataset1.append(-1 * np.array(force_stp_new))
    
```

```python
fig1, axes1 = plt.subplots(2, 2, sharey="row", sharex="col")
fig1.set_size_inches(13, 12)
for index, h in enumerate(h_lst1):

    axes1[0, 0].plot(velocity_lst1, force_stp_dataset1[index], linestyle='--', label='h = {}'.format(h))
    axes1[1, 0].plot(velocity_lst1, force_im_dataset1[index], linestyle='--')
    axes1[0, 1].plot(velocity_lst1, force_stp_new_dataset1[index], linestyle='--')
    axes1[1, 1].plot(velocity_lst1, force_im_new_dataset1[index], linestyle='--')


axes1.flat[0].set(ylabel='F_stopping')
axes1.flat[2].set(xlabel='Velocity (a.u)', ylabel='F_image')
axes1.flat[1].set(ylabel='F_stopping')
axes1.flat[3].set(xlabel='Velocity (a.u)', ylabel='F_image')

axes1[0, 0].set_title('Method from paper, gamma ~ 0 (0.1)')
axes1[0, 1].set_title('KKRs method with gamma = 0')

fig1.legend()
plt.show()
plt.close(fig1)


print('Done!')
```

```python
## Comparing the results between the original paper method and the new KKR method (fig 3): 

h2 = 8  # h in (a.u)
velocity_lst2 = np.linspace(0.01, 10, 30)
gamma_lst2 = [0.05, 0.1, 0.25, 0.5] 

force_stp_dataset2 = []
force_im_dataset2 = []
force_stp_new_dataset2 = []
force_im_new_dataset2 = []

for index, gamma in enumerate(gamma_lst2):
    print('processing gamma = {}'.format(gamma))
    force_stp = []
    force_im = []
    force_stp_new = []
    force_im_new = []

    for v in velocity_lst2:
        fstped = base.f_stp(v, gamma, h2)
        fim = base.f_im(v, gamma, h2)
        fstped_new = base.f_stp_kkr(v, gamma, h2)
        fim_new = base.f_im_kkr(v, gamma, h2)
        force_stp.append(fstped)
        force_im.append(fim)
        force_stp_new.append(fstped_new)
        force_im_new.append(fim_new)

    force_im_dataset2.append(-1 * np.array(force_im))
    force_stp_dataset2.append(-1 * np.array(force_stp))
    force_im_new_dataset2.append(-1 * np.array(force_im_new))
    force_stp_new_dataset2.append(-1 * np.array(force_stp_new))
```

```python
fig1, axes1 = plt.subplots(2, 2, sharey="row", sharex="col")
fig1.set_size_inches(13, 12)
for index, gamma in enumerate(gamma_lst2):
    axes1[0, 0].plot(velocity_lst2, force_stp_dataset2[index], linestyle='--', label='gamma = {}'.format(gamma))
    axes1[1, 0].plot(velocity_lst2, force_im_dataset2[index], linestyle='--')
    axes1[0, 1].plot(velocity_lst2, force_stp_new_dataset2[index], linestyle='--')
    axes1[1, 1].plot(velocity_lst2, force_im_new_dataset2[index], linestyle='--')


axes1.flat[0].set(ylabel='F_stopping')
axes1.flat[2].set(xlabel='Velocity (a.u)', ylabel='F_image')
axes1.flat[1].set(ylabel='F_stopping')
axes1.flat[3].set(xlabel='Velocity (a.u)', ylabel='F_image')

axes1[0, 0].set_title('Method from paper')
axes1[0, 1].set_title('KKRs Method')

fig1.legend()
plt.show()
plt.close(fig1)


print('Done!')
```
